### 0.)  What genres are associated with the highest median ROI?

```python
# To break up 'genre' column across a matrix, start with a new dataframe of empty columns for each genre.
genreList = []
for el in df_fulldata['genres'].map(lambda x: str(x).split(",")):
    for el2 in el:
        genreList.append(el2)
genreCols = list(set(genreList))
genres = pd.DataFrame(columns = genreCols)
```

```python
# Concatenate genres matrix with original dataset.
blankMatrix = pd.concat([df_fulldata, genres], axis = 1)
```

```python
# Create a function to fill NaNs in matrix. 
def creatematrix(df):
    for row_index, row in df.iterrows():
        for key, value in (dict(row)).items():
            if key == 'genres':
                for el in value.split(","):
                    df.loc[row_index, el] = 1
    df.fillna(0, inplace=True)
    return df

creatematrix(blankMatrix)
df_fulldata = blankMatrix
```

### 1.)  What genres groupings are associated with the highest median ROI?

```python
# Setting up a table with genre counts to inform percent of total by ROI category.
table6 = pd.pivot_table(df_fulldata, index=['genres'], values ='worldwide_roi', aggfunc=['mean','median','std','count'])
table6 = pd.DataFrame(table6)
table6.columns = table6.columns.to_series().str.join('_')
table6 = pd.DataFrame(table6.sort_values('median_worldwide_roi', ascending=False))
table6 = pd.DataFrame(table6.loc[table6['count_worldwide_roi']>9].reset_index())
table6.columns = ['genre_groupings', 'mean_worldwide_roi', 'median_worldwide_roi',
       'std_worldwide_roi', 'count_worldwide_roi']
table6[['genre_groupings','median_worldwide_roi','std_worldwide_roi']]
```

```python
# Creating a rank of genres across all ROI categories.
piv0 = pd.pivot_table(df_fulldata, index='genres', values='id', aggfunc='count')
piv0 = piv0.sort_values(by='id', ascending=False).reset_index()
piv0.index.name = 'genre_rank'
piv0 = piv0.reset_index()
piv0.columns = ['genreRank0', 'genres', 'count']
piv0
```

```python
# Setting up a table with genre counts to inform percent of total by ROI category.
piv1 = pd.pivot_table(df_fulldata, index=['roi_category','genres'], values ='worldwide_roi', aggfunc='count')
piv1.head()
```

```python
# Creating a new pivot table to use as divisor with the total count by ROI category, and performing division.
piv2 = piv1.groupby(['roi_category']).agg({'id': 'sum'})
piv3 = piv1.div(piv2, level='roi_category')*100
piv3.reset_index(inplace=True)
```

```python
# Create a ranking of ROI category to enable sorting.
piv3['r'] = piv3['roi_category'].apply(lambda x: {'high':0, 'somewhat high':1, 'moderate':2, 'somewhat low':3, 'low':4}[x])
piv3.columns = ['roi_category', 'genres', '%', 'r']
```

```python
# Adding genreRank0 to pivot table to enable sorting by top genres.
piv4 = piv3.merge(piv0, how="left", left_on='genres', right_on='genres').drop('count', axis=1)
piv4.sort_values(by=['r','genreRank0'], ascending=[True,True], inplace=True)
```

```python
# Add additional labels for visualization.
piv4['roi_category2'] = piv4['roi_category'].apply(lambda x: {'high':'high', 'somewhat high':'other', 'moderate':'other', 'somewhat low':'other', 'low':'other'}[x])
piv4['roi_category3'] = piv4['roi_category'].apply(lambda x: {'high':'high', 'somewhat high':np.nan, 'moderate':np.nan, 'somewhat low':np.nan, 'low':'low'}[x])
```

```python
piv4.head()
```

```python
# Limiting data for visualization.
piv5 = piv4.loc[piv4.genreRank0 < 5]
piv5.sort_values(by=['r','genreRank0'], ascending=[True,True], inplace=True)
```

```python
viz1 = sns.catplot(x='genres', y='%', hue='roi_category2', data=piv5, kind='bar', aspect=4);
viz1.set(ylabel="% of total by ROI Category");
viz1;
```

```python
plt.figure(figsize=(10, 3));
viz2 = sns.catplot(x='genres', y='%', hue='roi_category3', data=piv5, kind='bar', aspect=4);
viz2.set(ylabel="% of total by ROI Category");
sns.set_context('talk');
viz2;
```
